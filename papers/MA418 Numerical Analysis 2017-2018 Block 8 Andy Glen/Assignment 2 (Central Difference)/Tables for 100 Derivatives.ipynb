```python
from __future__ import division
from numpy import *
from sympy import *
import matplotlib.pyplot as plt
x, y, z, t = symbols('x y z t')
k, m, n = symbols('k m n', integer=True)
f, g, h = symbols('f g h', cls=Function)
import math

def f(x):
    return 1 - e**(-(x/2)**2)

def real_fprime(x):
    return (-x/2)*(-e**(-(x/2)**2))
    
def approx_fprime(x):
     return (f(x + h) - f(x))/h

def fprime_central(x):
    return (f(x +h) - f(x - h))/(2*h)   
```

```python
h = 0.001
```

```python
a = 0.05
print('   a        fp(x):forward    f(x):analytical  error:forward ')
print("      ")
while a < 5.00:
    print(' %.2f       %.10f     %.10f     %.10f   '%(a, approx_fprime(a), real_fprime(a), abs(approx_fprime(a) - real_fprime(a))))
    a = a + 0.05
```

```python
a = 0.05
print('  a         fp(x):central    f(x): analytical  error:central')
print("      ")
while a < 5.00:
    print(' %.2f       %.10f     %.10f      %.10f'%(a,  fprime_central(a), real_fprime(a), abs(fprime_central(a) - real_fprime(a))))
    a = a + 0.05    
```
