```python
def GaussTriple(a, b, m, n, p, cfn, dfn, alpha, beta, f):
    h1 = (b - a)/2
    h2 = (b + a)/2
    J = 0
    for i in range(1, m + 1):
        JX = 0
        x = h1*r[m-1][i-1] + h2
        d1 = dfn(x)
        c1 = cfn(x)
        k1 = (d1 - c1)/2
        k2 = (d1 + c1)/2
        for j in range(1, n + 1):
            JY = 0
            y = k1*r[n-1][j-1] + k2
            beta1 = beta(x,y)
            alpha1 = alpha(x,y)
            l1 = (beta1 - alpha1)/2
            l2 = (beta1 + alpha1)/2
            for k in range(1, p + 1):
                z = l1*r[p-1][k-1] + l2
                Q = f(x,y,z)
                JY = JY + c[p-1][k-1]*Q
            JX = JX + c[n-1][j-1]*l1*JY
        J = J + c[m-1][i-1]*k1*JX
    J = h1*J
    return J
```

```python
def c1(x):
    return 0
def d1(x):
    return 1 - x
def alpha(x,y):
    return x + y
def beta(x,y):
    return 3*x + 5*y
def f(x,y,z):
    return z

errorlist = []
for i in range(1, 11):
    n = i
    error = abs(23/12 - GaussTriple(0, 1, n, n, n, c1, d1, alpha, beta, f))
    errorlist = errorlist + [error]
```
